#### Prim's Algorithm

```python
# Algorthm :
# 1. Start from any random vertex (node 0)
# 2. Select the smallest edge that connects the selected vertex to a new vertex
# 3. Add that edge and the new vertex to the MST (Minimum Spanning Tree)
# 4. Repeat the process untill all vertices are included in the MST
```

```python
import matplotlib.pyplot as plt

V = 5
total_weight=0
G = [[0, 9, 75, 0, 0],
     [9, 0, 95, 19, 42],
     [75, 95, 0, 51, 66],
     [0, 19, 51, 0, 31],
     [0, 42, 66, 31, 0]]

selected = [False] * V
selected[0] = True

print("Edge : Weight")

for _ in range(V - 1):
    min_weight = float('inf')
    x, y = 0, 0

    for i in range(V):
        if selected[i]:
            for j in range(V):
                if not selected[j] and G[i][j] > 0:
                    if G[i][j] < min_weight:
                        min_weight = G[i][j]
                        
                        x, y = i, j
                        

    print(f"{x} - {y} : {G[x][y]}")
    total_weight=total_weight+min_weight
    selected[y] = True

print(f"Total cost of MST is {total_weight}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

V = 5
total_weight=0
G = [[0, 9, 75, 0, 0],
     [9, 0, 95, 19, 42],
     [75, 95, 0, 51, 66],
     [0, 19, 51, 0, 31],
     [0, 42, 66, 31, 0]]

selected = [False] * V
selected[0] = True

# For plotting
mst_edges = []

print("Edge : Weight")

for _ in range(V - 1):
    min_weight = float('inf')
    x, y = 0, 0

    for i in range(V):
        if selected[i]:
            for j in range(V):
                if not selected[j] and G[i][j] > 0:
                    if G[i][j] < min_weight:
                        min_weight = G[i][j]
                        
                        x, y = i, j
                        

    print(f"{x} - {y} : {G[x][y]}")
    total_weight=total_weight+min_weight
    selected[y] = True
    mst_edges.append((x, y, G[x][y]))

print(f"Total cost of MST is {total_weight}")


# Plotting the MST
G = nx.Graph()

# Add all vertices
for i in range(V):
    G.add_node(i)

# Add only MST edges
for u, v, w in mst_edges:
    G.add_edge(u, v, weight=w)

pos = nx.spring_layout(G)  # You can change layout to kamada_kawai_layout etc.

edge_labels = nx.get_edge_attributes(G, 'weight')

plt.figure(figsize=(8,6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=15, font_color="black", edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title("Minimum Spanning Tree using Prim's Algorithm")
plt.show()
```
