#### N-Queens

```python
# Objective: The objective of this lab assignment is to understand recursion and backtracking by implementing a Python program to solve the N-Queens problem.
# Problem Statement: The N-Queens problem involves placing N queens on an N Ã— N chessboard so that no two queens threaten each other. A queen can attack another queen if they are in the same row, column, or diagonal.
# Algorithm :
# 1. Create a board of size n x n filled with 0 cells
# 2. Place a queen in a row and check for conflicts
# 3. If safe, move to the next row; otherwise backtrack
# 4. Repeat untill all queens are placed successfully
```

```python
def solve_n_queens(n):
    def is_safe(board, row, col):
        for i in range(row):
            if board[i] == col or \
               board[i] - i == col - row or \
               board[i] + i == col + row:
                return False
        return True

    def queens(row, board, solutions):
        if row == n:
            solutions.append(board[:])
            return
        for col in range(n):
            if is_safe(board, row, col):
                board[row] = col
                queens(row + 1, board, solutions)
                board[row] = -1

    solutions = []
    board = [-1] * n
    queens(0, board, solutions)
    
    if solutions:
        print("\nUnique Solutions:")
        for sol in solutions:
            for i in range(n):
                print('.' * sol[i] + 'Q' + '.' * (n - sol[i] - 1))
            print()
    else:
        print("No solution exists")

if __name__ == "__main__":
    n = int(input("Enter the size of the chessboard: "))
    solve_n_queens(n)
```
